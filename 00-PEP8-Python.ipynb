#**Convenciones de nombres**

- **Princio primordial**

Los nombres que son visibles para el usuario como partes pblicas de la API deben seguir **convenciones que reflejen el uso en lugar de la implementacin**.

- **Descriptivo: Estilos de Nomenclatura**

Es til poder reconocer qu estilo de nomenclatura se est utilizando, independientemente de para qu se utilizan. 

Los siguientes estilos de nombres se distinguen comnmente:

*  1 sola letra minscula.
*  1 sola letra mayscula.
*  Palabra en minsculas.
*  Palabra en minsculas y guiones bajos.
*  Palabra en maysculas.
*  Palabra en maysculas y guiones bajos.
*  Palabras que se distinguen con mayscula inicial cada una de ellas(CapWords o CamelCase).
*  Palabras con combinacin maysculas y minsculas.
*  CapWords o CamnelCase con guiones bajos.

- **Prescriptivo: convenciones de nomenclatura**

  **Nombres a evitar**

  Nunca use los caracteres 'l' (letra minscula ele), 'O' (letra mayscula o) o 'I' (letra mayscula i) como nombres de variables de un solo carcter.

  **Compatibilidad ASCII**

  Los identificadores utilizados en la biblioteca estndar deben ser compatibles con ASCII, como se describe en la seccin de poltica de [PEP 3131](https://peps.python.org/pep-3131/).

  **Nombres de paquetes y mdulos**
  Los **mdulos** deben tener **nombres cortos y en minsculas**. Se pueden **usar guiones bajos** en el nombre del mdulo **si mejora la legibilidad**. Los **paquetes** de Python tambin deben tener nombres **cortos y en minsculas**, aunque se **desaconseja el uso de guiones bajos**.

  **Nombres de clases**

  Los nombres de las clases normalmente deberan usar la convencin CapWords.

  La convencin de nomenclatura para funciones se puede usar en los casos en que la interfaz est documentada y se usa principalmente como invocable.

  Tenga en cuenta que existe una convencin separada para los nombres incorporados (builtin names): la mayora de los nombres incorporados son palabras individuales (o dos palabras juntas), y la convencin CapWords se usa solo para nombres de excepcin y constantes incorporadas. Qu es un nombre incorporado? Son identificadores que Python emplea para uso dentro de la sintaxis en el lenguaje, no son **palabras reservadas**. 

  Para saber un poco ms sobre ello, puede consultar los siguientes enlaces:

  [Built-in Functions ](https://docs.python.org/3/library/functions.html#built-in-funcs)  
  [Built-in Constants](https://docs.python.org/3/library/constants.html#built-in-consts)

  **Nombres de variables de tipo**

  Los nombres de variables de tipo introducidos en [PEP 484](https://peps.python.org/pep-0484/) normalmente deben usar CapWords y prefieren nombres cortos: T, AnyStr, Num. Se recomienda agregar los sufijos _co o _contra a las variables utilizadas para declarar el comportamiento covariante o contravariante correspondientemente:

  ```

  VT_co = TypeVar('VT_co', covariant=True)
  KT_contra = TypeVar('KT_contra', contravariant=True)
  ```

  **Nombres de excepcin**

  **Nombres de variables globales**
  Se espera que stas variables estn pensadas para usarse dentro de un solo mdulo, las convenciones son ms o menos las mismas que las de las funciones.

  Los mdulos que estn diseados para su uso a travs de ```from M import * ```  deben usar el mecanismo  ```__all__``` para evitar la exportacin de globales.

  **Nombres de funciones y variables**

  Los nombres de las **funciones** deben estar en **minsculas**, con **palabras separadas por guiones bajos** segn sea necesario para **mejorar la legibilidad**.

  Los nombres de **variables** siguen la **misma convencin** que los nombres de **funciones**.

  **Argumentos de funcin y mtodo**

  Siempre se usa ```self``` para el primer argumento de los mtodos de instancia.

  Siempre se usa ```cls``` para el primer argumento de los mtodos de clase.

  Si el nombre de un argumento de funcin entra en conflicto con una palabra clave reservada, generalmente es mejor agregar un solo guin bajo en lugar de usar una abreviatura o corrupcin ortogrfica. Por lo tanto, class_ es mejor que clss, es mejor evitar tales conflictos usando un sinnimo.

  **Nombres de mtodos y variables de instancia**

  Use las reglas de nomenclatura de funciones: minsculas con palabras separadas por guiones bajos segn sea necesario para mejorar la legibilidad.

  Use un guin bajo inicial solo para mtodos no pblicos y variables de instancia.

  Para evitar conflictos de nombres con subclases, use dos guiones bajos iniciales para invocar las reglas de manipulacin de nombres de Python.

(Un usuario insistente an podra obtener acceso llamando a Foo._Foo__a).   **Diseo para la herencia**

  Siempre decida si los mtodos de una clase y las variables de instancia (colectivamente: "atributos") deben ser pblicos o no pblicos. En caso de duda, elija **no pblico; es ms fcil hacerlo pblico ms tarde que convertir un atributo pblico en no pblico**.

  Los atributos pblicos son aquellos que espera que utilicen los clientes no relacionados de su clase, con su compromiso de evitar cambios incompatibles con versiones anteriores. Los atributos no pblicos son aquellos que no estn destinados a ser utilizados por terceros; no garantiza que los atributos no pblicos no cambien o incluso se eliminen. **No se usa el trmino "privado"**, ya que **ningn atributo es realmente privado en Python**.

  Otra categora de atributos son aquellos que forman parte de la "subclase API" (a menudo denominada **"protegida"** en otros lenguajes). Algunas clases estn diseadas para ser heredadas, ya sea para ampliar o modificar aspectos del comportamiento de la clase. Al disear una clase de este tipo, tenga cuidado de tomar decisiones explcitas sobre qu atributos son pblicos, cules son parte de la API de la subclase y cules son realmente solo para su clase base.

  Ahora se enlistan algunas pautas para el diseo para la herencia:

  *  Los atributos pblicos no deben tener guiones bajos iniciales.
  *  Si el **nombre de su atributo pblico coincide con una palabra reservada, agregue un solo guin bajo al final de su nombre de atributo**.

  *  Si su clase est destinada a ser subclase y tiene atributos que no desea que usen las subclases, considere nombrarlos con guiones bajos dobles al principio y sin guiones bajos al final.

  **Interfaces pblicas e internas**

  Cualquier garanta de compatibilidad con versiones anteriores se aplica solo a las interfaces pblicas. En consecuencia, es importante que los usuarios puedan distinguir claramente entre las interfaces pblicas e internas.

  Las interfaces documentadas se consideran pblicas, a menos que la documentacin las declare explcitamente como interfaces provisionales o internas exentas de las garantas habituales de retrocompatibilidad. Se debe suponer que todas las interfaces no documentadas son internas.

  Una interfaz tambin se considera interna si cualquier espacio de nombres que la contenga (paquete, mdulo o clase) se considera interno.

  Los nombres importados siempre deben considerarse un detalle de implementacin. Otros mdulos no deben depender del acceso indirecto a tales nombres importados a menos que sean una parte explcitamente documentada de la API del mdulo contenedor, como os.path o el mdulo __init__ de un paquete que expone la funcionalidad de los submdulos.
