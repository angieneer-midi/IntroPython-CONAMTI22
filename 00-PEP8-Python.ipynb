#**Diseño de código**
---

- **Unir líneas de código**
  * **Explícitamente:** dos o más líneas físicas se pueden unir en líneas lógicas usando caracteres de barra invertida (\\).

  ``` 
   if 1900 < year < 2100 and 1 <= month <= 12 \
   and 1 <= day <= 31 and 0 <= hour < 24 \
   and 0 <= minute < 60 and 0 <= second < 60: 
   return 1
  ```

  - **Implícitamente:** las expresiones entre paréntesis, corchetes o llaves se pueden dividir en más de una línea física sin usar barras invertidas. 

  ```
  month_names = ['Januari', 'Februari', 'Maart',      # These are the
               'April',   'Mei',      'Juni',       # Dutch names
               'Juli',    'Augustus', 'September',  # for the months
               'Oktober', 'November', 'December']   # of the year
  ```

- **Identación**

  Python no utiliza llaves ```{ }``` para definir bloques de código, en su lugar emplea la **identación**, que en pocas palabras es el **"espacio"** que se deja a partir del margen izquierdo del editor donde se escribe nuestro código.

  Para este caso, se emplean 4 espacio por nivel de identación.

  Cuando se esbribe una línea de código y ésta se corta para continuar en la línea siguiente, todos los elementos se deben alinear de manera vertical, sobre todo lo que esté entre paréntesis, corchetes y llaves, o empleando una sangría francesa, si ésta se utiliza, no debe haber argumentos en la primera línea y se debe usar una sangría adicional para distinguirse claramente como continuación de línea.

  **¿Cómo hacerlo de forma correcta?**

```
# Elementos alineados con el delimitador de apertura 
foo = long_function_name(var_one, var_two,
                         var_three, var_four)

# Se agregan 4 espacios (se agrega un nivel extra de identación) para distinguir los argumentos del resto
def long_function_name(
        var_one, var_two, var_three,
        var_four):
    print(var_one)

# Se agrega un nivel de identación para los argumentos
foo = long_function_name(
    var_one, var_two,
    var_three, var_four)
```
  **¿Cómo no se debe aplicar la identación?**
```
#Los Argumentos de la primer línea no se encuentran alineados verticalmente
foo = long_function_name(var_one, var_two,
    var_three, var_four)

# Se requiere sangría adicional ya que la sangría no es distinguible.
def long_function_name(
    var_one, var_two, var_three,
    var_four):
    print(var_one)
```
La regla de los 4 espacios es opcional para continuar las líneas de código.

```
# Los niveles adicionales de identación pueden hacerse con 4 espacios.
foo = long_function_name(
  var_one, var_two,
  var_three, var_four)
```
Cuando la parte condicional de una declaración ```if``` es lo suficientemente larga como para requerir que se escriba en varias líneas. Es importante indicar que la combinación de la palabra reservada (2 caracteres), más un espacio, más el paréntesis que abre, todo ello crea una identación de 4 espacios de manera natural, permitiendo escribir una sentencia condicional multilínea. Sin embargo, esto puede producir un conflicto visual con el conjunto de código con sangría anidado dentro de la instrucción if, que naturalmente también tendría una sangría de 4 espacios. Las opciones aceptables en esta situación incluyen, pero no se limitan a:

```
# Sin alguna identación extra.
if (this_is_one_thing and
    that_is_another_thing):
    do_something()

if (this_is_one_thing and
    that_is_another_thing):
    # Comentario dentro del código
    do_something()

#Agregar identación extra a la línea de continuación condicional
if (this_is_one_thing
        and that_is_another_thing):
    do_something()
```
La llave/corchete/paréntesis de cierre en construcciones de varias líneas puede alinearse debajo del primer carácter que no sea un espacio en blanco de la última línea de la lista, como en:

```
my_list = [
    1, 2, 3,
    4, 5, 6,
    ]
result = some_function_that_takes_arguments(
    'a', 'b', 'c',
    'd', 'e', 'f',
    )
```
o puede alinearse bajo el primer carácter de la línea que inicia la construcción multilínea, como en:
```
my_list = [
    1, 2, 3,
    4, 5, 6,
]
result = some_function_that_takes_arguments(
    'a', 'b', 'c',
    'd', 'e', 'f',
)
```

  - **Palabras reservadas**

  Recordemos que, en todos los lenguajes de programación existen palabras que solo pueden emplearse para referirse a nombres de comandos, y en Python no es la excepción. Por lo tanto, éstas no es posible emplearlas para asignarlas como nombres (identificadores) a otros objetos. 

  Para conocer una parte del listado de éstas palabras puede ingresar el comando ```help('keywords')``` desde la interfaz interactiva de estas notas.

* **Soft keywords**

  Algunos **identificadores** solo se reservan en contextos específicos, tal como ```match, case y _``` pueden actuar sintácticamente como palabras clave en contextos relacionados con la declaración de coincidencia de patrones, pero esta distinción se realiza en el nivel del analizador, no cuando se tokeniza.

* **Clases reservadas de identificadores**
  
  Los nombres que empiezan con un carácter de subrayado están generalmente reservados para los nombres internos o privados. 
  Patrones de caracteres de subrayado inicial o final:
  * ```_*```  Proviene del módulo immport, e indica algo que no se importará.
  * ```_``` Cuando se encuentra en combinación de mayúsculas y minúsculas, coincide con una **soft keyword**.
  * ```__*__``` Se emplea para invocar métodos "mágicos" o también llamados dunder (doble underscores). Son utilizadas para crear funcionalidades que no pueden ser representadas en un método regular.
  * ```__*``` Se emplea para **Nombres privados de clase**, útiles para evitar conflictos de nombres entre los atributos "privados" de las clases base y derivadas.

- **Longitud máxima de una línea de código.**
  
  Limite todas las **líneas** a un máximo de **79 caracteres**.

  Para bloques de texto largos y fluidos con menos restricciones estructurales (**docstrings o comentarios**), la longitud de la línea debe limitarse a **72 caracteres**.

  Limitar el ancho requerido de la ventana del editor permite tener varios archivos abiertos uno al lado del otro y funciona bien cuando se usan herramientas de revisión de código que presentan las dos versiones en columnas adyacentes.

  La forma preferida de envolver líneas largas es usando la continuación de línea implícita de Python dentro de **paréntesis, corchetes y llaves**. Las **líneas largas se pueden dividir en varias líneas colocando expresiones entre paréntesis**. Estos deben usarse en lugar de usar una barra invertida para la continuación de la línea, por ejemplo, las sentencias with largas y múltiples no podían usar la continuación implícita antes de Python 3.10, por lo que las barras invertidas eran aceptables para ese caso:

```
with open('/path/to/some/file/you/want/to/read') as file_1, \
     open('/path/to/some/file/being/written', 'w') as file_2:
    file_2.write(file_1.read())
```
- **¿Debe romperse una línea antes o después de un operador binario?**

El estilo que se usa generalmente para romper una línea es después de los operadores binarios, pero al utilizarlo de esta forma la legibilidad de dos maneras: los operadores tienden a dispersarse en diferentes columnas en la pantalla, y cada operador se aleja de su operando y pasa a la línea anterior. En el siguiente ejemplo, es necesario revisar de manera adicional para saber qué elementos se suman y cuáles se restan:

```
# Forma errónea para cortar líneas con un operador binario
# los operadores se alejan de sus operandos
income = (gross_wages +
          taxable_interest +
          (dividends - qualified_dividends) -
          ira_deduction -
          student_loan_interest)
```
Donald Knuth explica la regla tradicional en su serie Computers and Typesetting: “Aunque las fórmulas dentro de un párrafo siempre se rompen después de las operaciones y relaciones binarias, las fórmulas mostradas siempre se rompen antes de las operaciones binarias” [3].

```
# Forma ideonea para hacer coincidir los operadores y operandos
income = (gross_wages
          + taxable_interest
          + (dividends - qualified_dividends)
          - ira_deduction
          - student_loan_interest)
```
- **Líneas en blanco**

  Considere colocar:

  * **Dos líneas** en blanco las **definiciones** de **funciones** y **clases de nivel superior**.
  * **Una línea** en blanco las **definiciones** de **métodos** dentro de una **clase**.
  * **Líneas en blanco** adicionales (con moderación) para **separar grupos de funciones relacionadas**.
  * **Líneas en blanco** en **funciones con moderación**, para indicar **secciones lógicas**.

- **Tipo de codificación del archivo fuente.**

  El código en la distribución principal de Python siempre debe usar **UTF-8 y no debe tener una declaración de codificación**.

  En la biblioteca estándar, las **codificaciones que no sean UTF-8** deben **usarse solo con fines de prueba**. Use caracteres que no sean ASCII con moderación, **preferiblemente solo para indicar lugares y nombres de personas**.

- **Importaciones (import)**
  * Las importaciones generalmente deben estar en líneas separadas:

  ```
  # Forma correcta:
  import os
  import sys

  # Forma incorrecta:
  import sys, os

  ```
  
  Sin embargo, esta forma de colocar una importación es correcta:

  ```
  from subprocess import Popen, PIPE
  ```

  * Las **importaciones** siempre se colocan en la **parte superior del archivo**, justo **después de los comentarios y cadenas de documentación** del módulo, pero **antes de las constantes y globales del módulo**.

    Las importaciones deben agruparse en el siguiente orden:

    1. Importaciones de biblioteca estándar.
    2. Importaciones de terceros relacionados.
    3. Importaciones específicas de bibliotecas/aplicaciones locales.

    Debe poner una línea en blanco entre cada grupo de importaciones.

  * Se recomiendan las importaciones absolutas, ya que suelen ser más legibles y tienden a comportarse mejor:

    ```
    import mypkg.sibling
    from mypkg import sibling
    from mypkg.sibling import example

    ```

- **Nombres a nivel de módulo para Dunders**

  Los "dunders" de nivel de módulo como __all__, __author__, __version__, etc. deben colocarse después de la cadena de documentación del módulo, pero antes de cualquier declaración de importación, excepto las importaciones ```from``` de ```__future__```. Python exige que las importaciones futuras deben aparecer en el módulo antes que cualquier otro código, excepto las cadenas de documentación:

  ```

  This module does stuff.
  """

  from __future__ import barry_as_FLUFL

  __all__ = ['a', 'b', 'c']
  __version__ = '0.1'
  __author__ = 'Cardinal Biggles'

  import os
  import sys
  ```


#**Manejo de cadenas**
---

En Python, las cadenas entre comillas simples y las cadenas entre comillas dobles son iguales.Cuando una cadena contiene caracteres de comillas simples o dobles, use la otra para evitar barras invertidas en la cadena. Mejora la legibilidad.

Para las cadenas entre comillas triples, utilice siempre caracteres de comillas dobles para ser coherente con la convención de cadenas de documentación que más adelante se analizará ese punto.

#**Espacios en blanco en expresiones y declaraciones**
---

Evitar espacios en blanco innnecesarios en las siguientes situaciones:
* Dentro de paréntesis, corchetes o llaves:

  ```
  # Correcto:
  spam(ham[1], {eggs: 2})
  ```

  ```
  # Incorrecto:
  spam( ham[ 1 ], { eggs: 2 } )
  ```

* Entre una coma final y un paréntesis de cierre siguiente:
  
  ```
  # Correcto:
  foo = (0,)
  ```

  ```
  # Incorrecto:
  bar = (0, )
  ```


* Inmediatamente antes de una coma, punto y coma o dos puntos:
  ```
  # Correcto:
  if x == 4: print(x, y); x, y = y, x
  ```
  
  ```
  # Incorrecto:
  if x == 4 : print(x , y) ; x , y = y , x
  ```

* Algunas ocasiones los dos puntos actúan como un operador binario y deben tener cantidades iguales en ambos lados (tratándolo como el operador con la prioridad más baja). Ambos dos puntos deben tener la misma cantidad de espacio aplicado. **Excepción:** cuando se omite un parámetro de segmento, se omite el espacio:
  
  ```
  # Correcto:
  ham[1:9], ham[1:9:3], ham[:9:3], ham[1::3], ham[1:9:]
  ham[lower:upper], ham[lower:upper:], ham[lower::step]
  ham[lower+offset : upper+offset]
  ham[: upper_fn(x) : step_fn(x)], ham[:: step_fn(x)]
  ham[lower + offset : upper + offset]
  ```

  ```
  # Incorrecto:
  ham[lower + offset:upper + offset]
  ham[1: 9], ham[1 :9], ham[1:9 :3]
  ham[lower : : upper]
  ham[ : upper]
  ```

* Antes del paréntesis abierto que inicia la lista de argumentos de una llamada de función:

  ```
  # Correcto:
  spam(1)
  ```

  ```
  # Incorrecto:
  spam (1)
  ```

* Inmediatamente antes del paréntesis abierto que inicia una indexación o división:

  ```
  # Correcto:
  dct['key'] = lst[index]
  ```

  ```
  # Incorrecto:
  dct ['key'] = lst [index]
  ```

* Más de un espacio alrededor de un operador de asignación (u otro) para alinearlo con otro:

  ```
  # Correct:
  x = 1
  y = 2
  long_variable = 3
  ```

  ```
  # Wrong:
  x             = 1
  y             = 2
  long_variable = 3
  ```

#**Comentarios**
---

  Son anotaciones que podemos realizar en el código, a modo que nos ayude a recordar información sobre el mismo. Para realizar esto en Python tenemos que ocupar el símbolo #, con ello definieremos comentarios en una sola línea. Si necesitamos hacer **comentarios en bloque** tendría que especificar varias líneas iniciando con el símbolo. O también puede realizar comentarios **en línea**, es decir, colocando delante de una línea de código un comentario.

  Los comentarios deben ser **oraciones completas**. La **primera palabra** debe estar en **mayúscula**, **a menos que sea un identificador ** que comience con una letra minúscula.

  Los **comentarios en bloque** generalmente consisten en **uno o más párrafos construidos a partir de oraciones completas**, y cada oración **termina en un punto**.

  Debe usar **dos espacios después de un punto final de oración en comentarios de varias oraciones**, excepto después de la oración final.

Asegúrese de que sus comentarios sean claros y fácilmente comprensibles para otros hablantes del idioma en el que está escribiendo.

Codificadores de Python de países de habla no inglesa: escriba sus comentarios en inglés, a menos que esté 120% seguro de que el código nunca será leído por personas que no hablen su idioma.

**Cadenas de documentación**
Las cadenas de documentación o **Docstring**, suelen ser considerados un tipo de comentario los cuales sirven para documentar algún módulo, clase, método, etc. 

En Python todos los objetos cuentan con una **variable** especial llamada **doc** gracias a la que podemos describir para qué sirven los y cómo se usan los objetos, por tanto, **todas las cadenas de documentación se convierten en el atributo __doc__** del objeto al que describe. La función interna de Python ```help(nombre)``` proporciona dicha documentación interna.

Si requiere ver más información sobre cómo especificar cadenas de documentación, puede consultarlo en la convención [PEP 257](https://peps.python.org/pep-0257/)


#**Convenciones de nombres**
---

- **Princio primordial**

Los nombres que son visibles para el usuario como partes públicas de la API deben seguir **convenciones que reflejen el uso en lugar de la implementación**.

- **Descriptivo: Estilos de Nomenclatura**

Es útil poder reconocer qué estilo de nomenclatura se está utilizando, independientemente de para qué se utilizan. 

Los siguientes estilos de nombres se distinguen comúnmente:

*  1 sola letra minúscula.
*  1 sola letra mayúscula.
*  Palabra en minúsculas.
*  Palabra en minúsculas y guiones bajos.
*  Palabra en mayúsculas.
*  Palabra en mayúsculas y guiones bajos.
*  Palabras que se distinguen con mayúscula inicial cada una de ellas(CapWords o CamelCase).
*  Palabras con combinación mayúsculas y minúsculas.
*  CapWords o CamnelCase con guiones bajos.

- **Prescriptivo: convenciones de nomenclatura**

  **Nombres a evitar**

  Nunca use los caracteres 'l' (letra minúscula ele), 'O' (letra mayúscula o) o 'I' (letra mayúscula i) como nombres de variables de un solo carácter.

  **Compatibilidad ASCII**

  Los identificadores utilizados en la biblioteca estándar deben ser compatibles con ASCII, como se describe en la sección de política de [PEP 3131](https://peps.python.org/pep-3131/).

  **Nombres de paquetes y módulos**
  Los **módulos** deben tener **nombres cortos y en minúsculas**. Se pueden **usar guiones bajos** en el nombre del módulo **si mejora la legibilidad**. Los **paquetes** de Python también deben tener nombres **cortos y en minúsculas**, aunque se **desaconseja el uso de guiones bajos**.

  **Nombres de clases**

  Los nombres de las clases normalmente deberían usar la convención CapWords.

  La convención de nomenclatura para funciones se puede usar en los casos en que la interfaz está documentada y se usa principalmente como invocable.

  Tenga en cuenta que existe una convención separada para los nombres incorporados (builtin names): la mayoría de los nombres incorporados son palabras individuales (o dos palabras juntas), y la convención CapWords se usa solo para nombres de excepción y constantes incorporadas. ¿Qué es un nombre incorporado? Son identificadores que Python emplea para uso dentro de la sintaxis en el lenguaje, no son **palabras reservadas**. 

  Para saber un poco más sobre ello, puede consultar los siguientes enlaces:

  [Built-in Functions ](https://docs.python.org/3/library/functions.html#built-in-funcs)  
  [Built-in Constants](https://docs.python.org/3/library/constants.html#built-in-consts)

  **Nombres de variables de tipo**

  Los nombres de variables de tipo introducidos en [PEP 484](https://peps.python.org/pep-0484/) normalmente deben usar CapWords y prefieren nombres cortos: T, AnyStr, Num. Se recomienda agregar los sufijos _co o _contra a las variables utilizadas para declarar el comportamiento covariante o contravariante correspondientemente:

  ```
  from typing import TypeVar

  VT_co = TypeVar('VT_co', covariant=True)
  KT_contra = TypeVar('KT_contra', contravariant=True)
  ```

  **Nombres de excepción**

  Debido a que las excepciones deben ser clases, aquí se aplica la convención de nomenclatura de clases. Sin embargo, debe usar el sufijo "Error" en los nombres de sus excepciones (si la excepción es realmente un error).
  
  **Nombres de variables globales**
  Se espera que éstas variables estén pensadas para usarse dentro de un solo módulo, las convenciones son más o menos las mismas que las de las funciones.

  Los módulos que están diseñados para su uso a través de ```from M import * ```  deben usar el mecanismo  ```__all__``` para evitar la exportación de globales.

  **Nombres de funciones y variables**

  Los nombres de las **funciones** deben estar en **minúsculas**, con **palabras separadas por guiones bajos** según sea necesario para **mejorar la legibilidad**.

  Los nombres de **variables** siguen la **misma convención** que los nombres de **funciones**.

  **Argumentos de función y método**

  Siempre se usa ```self``` para el primer argumento de los métodos de instancia.

  Siempre se usa ```cls``` para el primer argumento de los métodos de clase.

  Si el nombre de un argumento de función entra en conflicto con una palabra clave reservada, generalmente es mejor agregar un solo guión bajo en lugar de usar una abreviatura o corrupción ortográfica. Por lo tanto, class_ es mejor que clss, es mejor evitar tales conflictos usando un sinónimo.

  **Nombres de métodos y variables de instancia**

  Use las reglas de nomenclatura de funciones: minúsculas con palabras separadas por guiones bajos según sea necesario para mejorar la legibilidad.

  Use un guión bajo inicial solo para métodos no públicos y variables de instancia.

  Para evitar conflictos de nombres con subclases, use dos guiones bajos iniciales para invocar las reglas de manipulación de nombres de Python.

  Python mezcla estos nombres con el nombre de la clase: si la clase Foo tiene un atributo llamado __a, Foo.__a no puede acceder a él. (Un usuario insistente aún podría obtener acceso llamando a Foo._Foo__a). En general, los guiones bajos dobles deben usarse solo para evitar conflictos de nombres con atributos en clases diseñadas para ser subclasificadas.

  **Constantes**

  Las constantes generalmente se definen en un nivel de módulo y se escriben en letras mayúsculas con guiones bajos que separan las palabras. Los ejemplos incluyen MAX_OVERFLOW y TOTAL.

  **Diseño para la herencia**

  Siempre decida si los métodos de una clase y las variables de instancia (colectivamente: "atributos") deben ser públicos o no públicos. En caso de duda, elija **no público; es más fácil hacerlo público más tarde que convertir un atributo público en no público**.

  Los atributos públicos son aquellos que espera que utilicen los clientes no relacionados de su clase, con su compromiso de evitar cambios incompatibles con versiones anteriores. Los atributos no públicos son aquellos que no están destinados a ser utilizados por terceros; no garantiza que los atributos no públicos no cambien o incluso se eliminen. **No se usa el término "privado"**, ya que **ningún atributo es realmente privado en Python**.

  Otra categoría de atributos son aquellos que forman parte de la "subclase API" (a menudo denominada **"protegida"** en otros lenguajes). Algunas clases están diseñadas para ser heredadas, ya sea para ampliar o modificar aspectos del comportamiento de la clase. Al diseñar una clase de este tipo, tenga cuidado de tomar decisiones explícitas sobre qué atributos son públicos, cuáles son parte de la API de la subclase y cuáles son realmente solo para su clase base.

  Ahora se enlistan algunas pautas para el diseño para la herencia:

  *  Los atributos públicos no deben tener guiones bajos iniciales.
  *  Si el **nombre de su atributo público coincide con una palabra reservada, agregue un solo guión bajo al final de su nombre de atributo**.

    Nota 1: consulte la recomendación de nombre de argumento anterior para los métodos de clase.

  *  Para atributos de datos públicos simples, es mejor exponer solo el nombre del atributo, sin métodos complicados de acceso/mutación. Tenga en cuenta que Python proporciona un camino fácil para futuras mejoras, en caso de que descubra que un atributo de datos simple necesita desarrollar un comportamiento funcional. En ese caso, use propiedades para ocultar la implementación funcional detrás de la sintaxis de acceso a atributos de datos simples.

  *  Si su clase está destinada a ser subclase y tiene atributos que no desea que usen las subclases, considere nombrarlos con guiones bajos dobles al principio y sin guiones bajos al final.

  **Interfaces públicas e internas**

  Cualquier garantía de compatibilidad con versiones anteriores se aplica solo a las interfaces públicas. En consecuencia, es importante que los usuarios puedan distinguir claramente entre las interfaces públicas e internas.

  Las interfaces documentadas se consideran públicas, a menos que la documentación las declare explícitamente como interfaces provisionales o internas exentas de las garantías habituales de retrocompatibilidad. Se debe suponer que todas las interfaces no documentadas son internas.

  Una interfaz también se considera interna si cualquier espacio de nombres que la contenga (paquete, módulo o clase) se considera interno.

  Los nombres importados siempre deben considerarse un detalle de implementación. Otros módulos no deben depender del acceso indirecto a tales nombres importados a menos que sean una parte explícitamente documentada de la API del módulo contenedor, como os.path o el módulo __init__ de un paquete que expone la funcionalidad de los submódulos.
