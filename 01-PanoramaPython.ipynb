## Diccionarios

Los diccionarios en python son un tipo de dato utilizado para almacenar pares de "clave: valor". Cada clave en un diccionario tendr que ser nica y deber utilizar un tipo de dato inmutable (strings y datos numricos). Los valores asociados a una clave pueden ser cualquier tipo de dato en python o una combinacin de stos. Un diccionario es una secuencia no ordenada de elementos.

Un diccionario representa un mapeo desde las claves definidas haca los valores. Cada clave nica mapea a un valor. Como en un diccionario de ingls a espaol.

Para definir un diccionario, se utiliza ```{}```. Cada par ```clave:valor``` deber estar separado por ```,```.

```
# un diccionario que mapea nombres(claves) a numeros(valores).
contacts = {"jack": 1, "ellie":2, "charles":3, "kelly":4} 
```

Para acceder a los valores de un diccionario, se utiliza la clave como ndice.

```
#Acceder al valor de jack en el diccionario
print(contacts["jack"])
```

Si se usa una clave que no est en el diccionario recibir un error.

Es posible modificar el valor asociado a una clave usando la clave como ndice.

```
#asignando un nuevo valor a kelly
contacts["kelly"] = 0
```

Tambin es posible agregar nuevos pares clave valor simplemente asignandolos a un diccionario existente.

```
# Agregando un nuevo par clave:valor
contacts["jean"] = 5
```

````python
#Un diccionario que mapea nombres a numeros
contacts = {"jack": 1, "ellie":2, "charles":3, "kelly":4} 
print(contacts)

#acceder a un valor especifico usando la clave como indice
print(contacts["jack"])

#modificar el valor asignado a un elemento del diccionario usando la clave
contacts["kelly"] = 0
print(contacts)

#agregando un nuevo elemento clave:valor al diccionario
contacts["jean"] = 5
print(contacts)
````

````python
contacts = {"jack": 1, "ellie":2, "charles":3, "kelly":4} 

#intentando acceder al valor de una clave que no esta en el diccionario
print(contacts["jean"])
````

En python, la primera parte de una sentencia condicional est formada por la palabra reservada ```if``` seguida de una expresin booleana. ## Operadores relacionales
Los siguientes operadores relacionales pueden ser utilizados para realizar comparaciones entre variables en una sentencia condicional.


* ```x == y``` equivalente a: True si el valor de $x$ es igual al de $y$

* ```x != y``` diferente a: True si el valor de $x$ es diferente al de $y$

* ```x < y``` menor a: True si el valor de $x$ es menor al de $y$

* ```x > y``` mayor a: True si el valor de $x$ es mayor al de $y$

* ```x <= y``` # menor o igual a: True si el valor de $x$ es menor o igual al de $y$

* ```x >= y``` mayor o igual a: True si el valor de $x$ es mayor o igual al de $y$


````python
#Intenta cambiar los valores de x, y
x = 5
y = 5

print("x == y", x == y)
print("x != y", x != y)
print("x < y", x < y)
print("x > y", x > y)
print("x <= y", x <= y)
print("x >= y", x >= y)
````

# Ciclos

Muchas veces en un programa es necesario ejecutar una secuencia de cdigo repetidamente para resolver una tarea. Los lenguajes de programacin permiten hacer esto mediante ciclos, que ejecutan cdigo mltiples ocasiones hasta que se cumple una condicin de paro.

## While

El ciclo ```while``` permite ejecutar una secuencia de cdigo repetidamente mientras la expresin booleana recibida sea verdadera. La estructura de un ciclo while en python es la siguiente:

```
while expresion_booleana:
  # ejecuta secuencia de cdigo
```

_Nota: Un ciclo ```while``` podra ejecutarse sin parar si la expresin booleana siempre es verdadera (y nunca cambiara)._


````python
x = 1

#intenta cambiar la expresion booleana
while x < 10:
    print(x)
    #reasignando a x el valor actual de x mas 1
    x = x + 1
````

## For

El ciclo ```for``` es la segunda opcin de para realizar ciclos. En python, ```for``` es utilizado para recorrer _objetos_ iterables (como una lista) y ejecutar una secuencia de cdigo para cada elemento de sta.

```
for el in lista:
  # secuencia de codigo
```

````python
lista = [1, 2, 3, 4, 5]

# recorriendo la lista elemento por elemento
print("Usando for para recorrer e imprimir los elementos de una lista")
for el in lista:
    print(el)
````

El tipo de dato range tambin se convierte til en ciclos ```for```, ya que podemos recorrer un rango de nmeros durante un ciclo ```for```.

```
for i in range(10):
  # secuencia de cdigo
```

````python
# Prueba a cambiar el valor de n
n = 5
# usando for y range para recorrer un rango de numeros
print("Usando for y range para imprimir n numeros")
for i in range(n):
    print(i)
````

### List comprehensions

En python, existe una manera compacta de crear listas conocida como list comprehession. Para esto usas un for dentro de la declaracin de una lista para definir los elementos. Por ejemplo:

```
l = [x_exp for x in coleccion]
```

````python
# generando una lista con 10 numeros con un for
numbers = []

for x in range(10):
    numbers.append(x)
print(numbers)

# usando list comprehession
numbers = [x for x in range(10)]
print(numbers)
````

````python
# Algo mas avanzado
# una lista con los cuadrados de los numeros de 0 al 9
squares = []
for x in range(10):
    squares.append(x*x)
print(squares)
  
# usando list comprehession
squares = [x*x for x in range(10)]
print(squares)
````

**Condicional en list comprehession**

Tambin es posible usar listas ms avanzadas con list comprehession usando condicionales.

````python
# una lista con los cuadrados de los numeros pares entre 0 a 20

even_squares = [x*x for x in range(20) if x%2 == 0]
print(even_squares)
````

# Mdulos

Python es un lenguaje que permite trabajar ms de un paradigma (programacin estructurada y orientada a objetos). Generalmente cuando se codifica un programa aplicamos el paradigma de programacin estructurada (Secuencia, seleccin e iteracin), y cuando necesitamos programar un proyecto un poco ms complejo y queremos incluir los beneficios del paradigma orientado a objetos (reutilizacin de los componentes, facilidad de mantenimiento y modificacin de objetos) se utilizan clases (POO). 

Dividir un programa permite reutilizar procesos o **funciones** que se escribieron con anterioridad en un nuevo programa sin tener que copiar el cdigo. Esto es, escribir definiciones en un archivo y usarlas durante otro programa. A este tipo de archivos se le conoce como **mdulos**. Las definiciones en un mdulo puede ser "importadas" a otros mdulos o a un programa principal.

El archivo de un mdulo llevar como nombre el nombre del mdulo concatenado con el sufijo ```.py```. El archivo contendr las respectivas definiciones y declaraciones.

Python por defecto incluye una librera de mdulos, conocidos como mdulos estndar. Adems, algunas plataformas de python contienen una gran variedad de mdulos instalados por defecto. Los mdulos permite extender las funcionalidades de python y llevar a cabo una gran variedad de tareas.

## Importando mdulos estndar

Python provee una librera de mdulos estndar que puedes importar y utilizar. Por ejemplo, el mdulo math contiene mltiples funciones matemticas.

Para utilizar los mdulos, debes importarlos usando la palabra ```import``` seguida del nombre del mdulo:


```

```


````python
#Import el modulo math
import math
````

Al importar un mdulo se puede utilizar un **alias** para cambiar el nombre con el que lo llamars durante el cdigo, y de esta forma que ste sea ms fcil de identificarlo. Para asignar un alias, utiliza la palabra ```as``` seguida del alias. Una vez que se ha importado un mdulo con un alias, se debe usar el alias en el cdigo.
